<b>Calculate Lipinski descriptors:</b>
Rule for evaluating the druglikeness of compounds. Absorption, Distribution, Metabolism, and Excretion (ACME) or the pharmacokinetic profile. All orally active FDA-approved drugs are in form of <b>Lipinski's Rule</b>:

<li>molecular weight < 500 Daltons</li> 
<li>Octanol-water partition coefficient (LogP) < 5</li>
<li>Hydrogen bond donors < 5</li>
<li>Hydrogen bond acceptors <>> 10</li>

```python
'''Import libraries'''
import numpy as np
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski
```

```python
'''Define function to calculate Lipinski Descriptors'''
def lipinski(smiles, verbose=False):

    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem) 
        moldata.append(mol)
       
    baseData= np.arange(1,1)
    i=0  
    for mol in moldata:        
       
        desc_MolWt = Descriptors.MolWt(mol)
        desc_MolLogP = Descriptors.MolLogP(mol)
        desc_NumHDonors = Lipinski.NumHDonors(mol)
        desc_NumHAcceptors = Lipinski.NumHAcceptors(mol)
           
        row = np.array([desc_MolWt,
                        desc_MolLogP,
                        desc_NumHDonors,
                        desc_NumHAcceptors])   
    
        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1      
    
    columnNames=["MW","LogP","NumHDonors","NumHAcceptors"]   
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)
    
    return descriptors
```

```python
'''Call on dataframe'''
df_lipinski = lipinski(df.canonical_smiles)
df_lipinski
```

```python
'''Combine dataframes'''
df_combined = pd.concat([df, df_lipinski], axis=1)
df_combined
```

<b>Convert IC50 to pIC50</b>
-log10(IC50), pIC50, creates a more uniform distribution. Accepts a dataframe as input then will:

<li>Take the IC50 values from the ``standard_value`` column and converts it from nM to M by multiplying the value by 10$^{-9}$</li> 
<li>Take the molar value and apply -log10</li>
<li>Delete the ``standard_value`` column and create a new ``pIC50`` column</li>

```python
'''Define conversion function'''
def pIC50(df):
    pIC50 = []

    for i in df['standard_value_norm']:
        molar = i*(10**-9) # Converts nM to M
        pIC50.append(-np.log10(molar))

    df['pIC50'] = pIC50
    x = df.drop('standard_value_norm', 1)
        
    return x
```

```python
'''Get the largest number. Cap to 100 million'''
df_combined.standard_value.describe()
```

```python
'''Define normalization function'''
def norm_value(df):
    norm = []

    for i in df['standard_value']:
        if i > 100000000:
          i = 100000000
        norm.append(i)

    df['standard_value_norm'] = norm
    x = df.drop('standard_value', 1)
        
    return x
```

```python
'''Call on dataframe'''
df_norm = norm_value(df_combined)
df_norm
```

```python
df_norm.standard_value_norm.describe()
```

```python
'''Call pIC50 conversion function'''
df_final = pIC50(df_norm)
df_final
```

Mann-Whitney U Test: Analyze statistical significance between bioactivity classes

```python
'''Retrieved from
  # https://machinelearningmastery.com/nonparametric-statistical-significance-tests-in-python/
'''
def mannwhitney(descriptor, verbose=False):

  from numpy.random import seed
  from numpy.random import randn
  from scipy.stats import mannwhitneyu

# seed the random number generator
  seed(1)

# actives and inactives
  selection = [descriptor, 'bioactivity']
  df = df2[selection]
  active = df[df['bioactivity'] == 'active']
  active = active[descriptor]

  selection = [descriptor, 'bioactivity']
  df = df2[selection]
  inactive = df[df['bioactivity'] == 'inactive']
  inactive = inactive[descriptor]

# compare samples
  stat, p = mannwhitneyu(active, inactive)
  #print('Statistics=%.3f, p=%.3f' % (stat, p))

# interpret
  alpha = 0.05
  if p > alpha:
    interpretation = 'Same distribution (fail to reject H0)'
  else:
    interpretation = 'Different distribution (reject H0)'
  
  results = pd.DataFrame({'Descriptor':descriptor,
                          'Statistics':stat,
                          'p':p,
                          'alpha':alpha,
                          'Interpretation':interpretation}, index=[0])
  filename = 'mannwhitneyu_' + descriptor + '.csv'
#   results.to_csv(filename)

  return results
```

```python
'''MannWhitney test: pIC50'''
mannwhitney('pIC50')
```
